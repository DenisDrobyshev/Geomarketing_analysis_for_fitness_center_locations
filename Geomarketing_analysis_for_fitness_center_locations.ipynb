```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point, Polygon
import osmnx as ox
import h3
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import folium
import matplotlib.pyplot as plt
import seaborn as sns

ox.settings.log_console = True
ox.settings.use_cache = True

RUSSIAN_FEATURE_NAMES = {
    'transport_within_500m': 'Транспорт в радиусе 500м',
    'commerce_within_500m': 'Торговля в радиусе 500м',
    'education_within_500m': 'Образование в радиусе 500м',
    'residential_within_500m': 'Жилье в радиусе 500м',
    'office_within_500m': 'Офисы в радиусе 500м',
    'leisure_within_500m': 'Досуг в радиусе 500м',
    'transport_within_1000m': 'Транспорт в радиусе 1000м',
    'commerce_within_1000m': 'Торговля в радиусе 1000м',
    'education_within_1000m': 'Образование в радиусе 1000м',
    'residential_within_1000m': 'Жилье в радиусе 1000м',
    'office_within_1000m': 'Офисы в радиусе 1000м',
    'leisure_within_1000m': 'Досуг в радиусе 1000м',
    'distance_to_nearest_transport': 'Расст. до транспорта',
    'distance_to_nearest_commerce': 'Расст. до торговли',
    'distance_to_nearest_education': 'Расст. до образования',
    'distance_to_nearest_residential': 'Расст. до жилья',
    'distance_to_nearest_office': 'Расст. до офисов',
    'distance_to_nearest_leisure': 'Расст. до досуга',
    'transport_within_500m_density': 'Плотность транспорта',
    'commerce_within_500m_density': 'Плотность торговли',
    'education_within_500m_density': 'Плотность образования',
    'residential_within_500m_density': 'Плотность жилья',
    'office_within_500m_density': 'Плотность офисов',
    'leisure_within_500m_density': 'Плотность досуга'
}

class GeoDataCollector:
    def __init__(self, city_name="Москва", crs="EPSG:4326"):
        self.city_name = city_name
        self.crs = crs
        self.gdf_dict = {}

    def get_city_boundary(self):
        """Получение границ города"""
        try:
            self.city_gdf = ox.geocode_to_gdf(self.city_name)
            self.city_gdf = self.city_gdf.to_crs(self.crs)
            print(f"Границы города {self.city_name} загружены")
            return self.city_gdf
        except Exception as e:
            print(f"Ошибка загрузки границ города: {e}")
            return self._create_demo_boundary()

    def _create_demo_boundary(self):
        """Создание демонстрационных границ"""
        lat, lon = 55.7558, 37.6173
        boundary = Polygon([
            [lon-0.1, lat-0.05], [lon+0.1, lat-0.05],
            [lon+0.1, lat+0.05], [lon-0.1, lat+0.05],
            [lon-0.1, lat-0.05]
        ])
        return gpd.GeoDataFrame({'name': ['demo_area']}, geometry=[boundary], crs="EPSG:4326")

    def get_osm_data(self, tags, name):
        """Загрузка данных из OSM по тегам"""
        try:
            gdf = ox.features_from_polygon(self.city_gdf.geometry.iloc[0], tags=tags)
            gdf = gdf.reset_index()
            gdf = gdf[gdf.geometry.notnull()]

            if not isinstance(gdf, gpd.GeoDataFrame):
                gdf = gpd.GeoDataFrame(gdf, geometry='geometry')

            gdf = gdf.set_geometry('geometry').to_crs(self.crs)
            self.gdf_dict[name] = gdf
            print(f"Загружено {len(gdf)} объектов: {name}")
            return gdf
        except Exception as e:
            print(f"Ошибка загрузки {name}: {e}")
            return gpd.GeoDataFrame(columns=['geometry'], crs=self.crs)

    def load_all_data(self):
        """Загрузка всех необходимых данных"""
        print("Загрузка данных инфраструктуры...")

        # Фитнес-центры
        fitness_tags = {'leisure': 'fitness_centre', 'amenity': 'gym', 'sport': 'fitness'}
        self.get_osm_data(fitness_tags, "fitness_centers")

        # Инфраструктура
        infrastructures = [
            ({'railway': ['station', 'tram_stop'], 'highway': ['bus_stop']}, "transport"),
            ({'shop': True, 'amenity': ['bank', 'pharmacy', 'cafe', 'restaurant']}, "commerce"),
            ({'amenity': ['school', 'university', 'college']}, "education"),
            ({'building': ['apartments', 'residential', 'dormitory']}, "residential"),
            ({'building': ['office', 'commercial']}, "office"),
            ({'leisure': ['park', 'sports_centre', 'pitch']}, "leisure")
        ]

        for tags, name in infrastructures:
            self.get_osm_data(tags, name)

        print("Итоги загрузки данных:")
        for name, gdf in self.gdf_dict.items():
            print(f"   {name}: {len(gdf)} объектов")

class HexGridBuilder:
    def __init__(self, resolution=9):
        self.resolution = resolution

    def create_hex_grid(self, boundary_gdf):
        """Создание гексагональной сетки H3"""
        try:
            boundary_wgs84 = boundary_gdf.to_crs("EPSG:4326").unary_union
            bounds = boundary_wgs84.bounds
            min_lon, min_lat, max_lon, max_lat = bounds

            hexagons = set()
            lats = np.linspace(min_lat, max_lat, 50)
            lons = np.linspace(min_lon, max_lon, 50)

            for lat in lats:
                for lon in lons:
                    if boundary_wgs84.contains(Point(lon, lat)):
                        hexagons.add(h3.latlng_to_cell(lat, lon, self.resolution))

            geometries, hex_ids = [], []
            for hex_id in hexagons:
                try:
                    hex_boundary = h3.cell_to_boundary(hex_id)
                    polygon = Polygon([(lon, lat) for lat, lon in hex_boundary])
                    geometries.append(polygon)
                    hex_ids.append(hex_id)
                except Exception:
                    continue

            hex_gdf = gpd.GeoDataFrame({'hex_id': hex_ids, 'geometry': geometries}, crs="EPSG:4326")
            hex_gdf = hex_gdf.to_crs(boundary_gdf.crs)
            print(f"Создано {len(hex_gdf)} гексагонов")
            return hex_gdf

        except Exception as e:
            print(f"Ошибка создания сетки: {e}")
            return self._create_demo_hex_grid(boundary_gdf)

    def _create_demo_hex_grid(self, boundary_gdf):
        """Создание демонстрационной сетки"""
        bounds = boundary_gdf.bounds.iloc[0]
        minx, miny, maxx, maxy = bounds.minx, bounds.miny, bounds.maxx, bounds.maxy
        hex_size = (maxx - minx) / 20

        geometries, hex_ids = [], []
        for i in range(10):
            for j in range(10):
                x = minx + i * hex_size * 1.5
                y = miny + j * hex_size * np.sqrt(3)
                hexagon = Polygon([
                    [x, y], [x + hex_size/2, y + hex_size * np.sqrt(3)/2],
                    [x + hex_size * 1.5, y + hex_size * np.sqrt(3)/2], [x + 2 * hex_size, y],
                    [x + hex_size * 1.5, y - hex_size * np.sqrt(3)/2],
                    [x + hex_size/2, y - hex_size * np.sqrt(3)/2], [x, y]
                ])
                geometries.append(hexagon)
                hex_ids.append(f"demo_{i}_{j}")

        demo_gdf = gpd.GeoDataFrame({'hex_id': hex_ids, 'geometry': geometries}, crs=boundary_gdf.crs)
        print(f"Создано {len(demo_gdf)} демо-гексагонов")
        return demo_gdf

class FeatureEngineer:
    def __init__(self, hex_gdf, infrastructure_data):
        self.hex_gdf = hex_gdf
        self.infrastructure_data = infrastructure_data
        self.features_df = pd.DataFrame({'hex_id': hex_gdf['hex_id']})

    def calculate_spatial_features(self, radii=[500, 1000]):
        """Расчет всех пространственных признаков"""
        print("Расчет пространственных признаков...")

        # Признаки в радиусе
        for radius in radii:
            for feature_type, gdf in self.infrastructure_data.items():
                if feature_type != "fitness_centers":
                    self._calculate_features_in_radius(gdf, feature_type, radius)

        # Расстояния до ближайших объектов
        for feature_type, gdf in self.infrastructure_data.items():
            if feature_type != "fitness_centers":
                self._calculate_distance_to_nearest(gdf, feature_type)

        # Плотностные характеристики
        self._calculate_density_features()

        print(f"Создано {len(self.features_df.columns) - 1} признаков")
        return self

    def _calculate_features_in_radius(self, gdf, feature_name, radius=500):
        """Расчет количества объектов в радиусе"""
        counts = []
        hex_gdf_crs = self.hex_gdf.to_crs('EPSG:3857')

        if not gdf.empty:
            gdf_crs = gdf.to_crs('EPSG:3857')
            for _, hex_row in hex_gdf_crs.iterrows():
                buffer = hex_row.geometry.centroid.buffer(radius)
                counts.append(len(gdf_crs[gdf_crs.geometry.within(buffer)]))
        else:
            counts = [0] * len(self.hex_gdf)

        self.features_df[f'{feature_name}_within_{radius}m'] = counts

    def _calculate_distance_to_nearest(self, gdf, feature_name):
        """Расчет расстояния до ближайшего объекта"""
        if gdf.empty:
            self.features_df[f'distance_to_nearest_{feature_name}'] = [999999] * len(self.hex_gdf)
            return

        hex_gdf_crs = self.hex_gdf.to_crs('EPSG:3857')
        gdf_crs = gdf.to_crs('EPSG:3857')

        distances = []
        for _, hex_row in hex_gdf_crs.iterrows():
            dist = gdf_crs.geometry.distance(hex_row.geometry.centroid).min()
            distances.append(dist)

        self.features_df[f'distance_to_nearest_{feature_name}'] = distances

    def _calculate_density_features(self):
        """Расчет плотностных характеристик"""
        self.features_df['hex_area_sqkm'] = self.hex_gdf.to_crs('EPSG:3857').geometry.area / 1000000

        for col in self.features_df.columns:
            if 'within' in col and 'density' not in col:
                self.features_df[f'{col}_density'] = (
                    self.features_df[col] / self.features_df['hex_area_sqkm']
                ).fillna(0)

    def normalize_features(self):
        """Нормализация признаков"""
        feature_cols = [col for col in self.features_df.columns
                       if col not in ['hex_id', 'hex_area_sqkm', 'target']]

        self.features_df[feature_cols] = self.features_df[feature_cols].fillna(0)
        scaler = StandardScaler()
        self.features_df[feature_cols] = scaler.fit_transform(self.features_df[feature_cols])
        self.scaler = scaler
        return self

class FitnessLocationModel:
    def __init__(self):
        self.model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

    def prepare_target_variable(self, features_df, fitness_gdf, hex_gdf):
        """Формирование целевой переменной"""
        if fitness_gdf.crs != hex_gdf.crs:
            fitness_gdf = fitness_gdf.to_crs(hex_gdf.crs)

        target = []
        for _, hex_row in hex_gdf.iterrows():
            fitness_in_hex = fitness_gdf[fitness_gdf.geometry.within(hex_row.geometry)]
            target.append(1 if len(fitness_in_hex) > 0 else 0)

        features_df['target'] = target
        print(f"Целевая переменная создана: {sum(target)} положительных примеров")
        return features_df

    def train(self, X, y):
        """Обучение модели"""
        self.feature_cols = [col for col in X.columns
                           if col not in ['hex_id', 'target', 'hex_area_sqkm']]
        X_train = X[self.feature_cols]

        # Проверка наличия обоих классов
        if len(np.unique(y)) < 2:
            print("Только один класс в целевой переменной")
            y = y.copy()
            y.iloc[:5] = 1 - y.iloc[0]

        self.model.fit(X_train, y)

        # Кросс-валидация
        try:
            cv_scores = cross_val_score(self.model, X_train, y, cv=3, scoring='roc_auc', n_jobs=-1)
            print(f"Random Forest CV ROC-AUC: {cv_scores.mean():.3f} (±{cv_scores.std() * 2:.3f})")
        except Exception as e:
            print(f"Кросс-валидация не удалась: {e}")

        return self

    def analyze_feature_importance(self):
        """Анализ важности признаков"""
        if hasattr(self.model, 'feature_importances_'):
            importance_df = pd.DataFrame({
                'feature': self.feature_cols,
                'importance': self.model.feature_importances_
            }).sort_values('importance', ascending=False)

            importance_df['feature_russian'] = importance_df['feature'].map(
                lambda x: RUSSIAN_FEATURE_NAMES.get(x, x)
            )

            return importance_df
        return None

class TerritoryClusterer:
    def __init__(self, max_clusters=8):
        self.max_clusters = max_clusters

    def find_optimal_clusters(self, X, max_samples=1000):
        """Определение оптимального числа кластеров"""
        feature_cols = [col for col in X.columns
                       if col not in ['hex_id', 'target', 'hex_area_sqkm']]
        X_cluster = X[feature_cols].fillna(0)

        if len(X_cluster) > max_samples:
            X_sample = X_cluster.sample(max_samples, random_state=42)
        else:
            X_sample = X_cluster

        cluster_range = range(2, min(self.max_clusters + 1, len(X_sample)))
        wcss, silhouette_scores = [], []

        for n_clusters in cluster_range:
            kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
            cluster_labels = kmeans.fit_predict(X_sample)
            wcss.append(kmeans.inertia_)
            if n_clusters > 1:
                silhouette_scores.append(silhouette_score(X_sample, cluster_labels))

        # Визуализация
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        ax1.plot(cluster_range, wcss, marker='o')
        ax1.set_title('Метод локтя')
        ax1.set_xlabel('Количество кластеров')
        ax1.set_ylabel('WCSS')

        ax2.plot(cluster_range, silhouette_scores, marker='o')
        ax2.set_title('Silhouette Score')
        ax2.set_xlabel('Количество кластеров')
        ax2.set_ylabel('Score')

        plt.tight_layout()
        plt.show()

        optimal_clusters = cluster_range[np.argmax(silhouette_scores)] if silhouette_scores else 3
        print(f"Оптимальное число кластеров: {optimal_clusters}")
        return optimal_clusters

    def perform_clustering(self, X, n_clusters):
        """Выполнение кластеризации"""
        feature_cols = [col for col in X.columns
                       if col not in ['hex_id', 'target', 'hex_area_sqkm']]
        X_cluster = X[feature_cols].fillna(0)

        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(X_cluster)

        return cluster_labels, kmeans

class PotentialCalculator:
    def __init__(self, attractiveness_weight=0.7, competition_weight=0.3):
        self.weights = (attractiveness_weight, competition_weight)

    def calculate(self, hex_gdf, features_df, model, fitness_gdf):
        """Расчет интегрального потенциала"""
        X = features_df[model.feature_cols].copy().fillna(0)
        attractiveness = model.model.predict_proba(X)[:, 1]

        # Расчет конкуренции
        hex_gdf_metric = hex_gdf.to_crs('EPSG:3857')
        if not fitness_gdf.empty:
            fitness_gdf_metric = fitness_gdf.to_crs('EPSG:3857')

        competition_scores = []
        for _, hex_row in hex_gdf_metric.iterrows():
            buffer = hex_row.geometry.centroid.buffer(500)
            if not fitness_gdf.empty:
                competitors = fitness_gdf_metric[fitness_gdf_metric.geometry.within(buffer)]
                competition_scores.append(len(competitors))
            else:
                competition_scores.append(0)

        # Нормализация конкуренции
        if competition_scores and max(competition_scores) > 0:
            max_comp = max(competition_scores)
            normalized_competition = [1 - (score / max_comp) for score in competition_scores]
        else:
            normalized_competition = [1] * len(competition_scores)

        integral_potential = (
            self.weights[0] * attractiveness +
            self.weights[1] * np.array(normalized_competition)
        )

        return integral_potential, attractiveness, normalized_competition

    def categorize(self, potential_scores, n_categories=5):
        """Категоризация потенциала"""
        potential_series = pd.Series(potential_scores)
        unique_values = len(potential_series.unique())
        actual_categories = min(n_categories, unique_values)

        if actual_categories <= 1:
            return ['Очень высокий'] * len(potential_scores)

        category_names = {
            1: ['Низкий'], 2: ['Низкий', 'Высокий'],
            3: ['Низкий', 'Средний', 'Высокий'],
            4: ['Очень низкий', 'Низкий', 'Высокий', 'Очень высокий'],
            5: ['Очень низкий', 'Низкий', 'Средний', 'Высокий', 'Очень высокий']
        }

        labels = category_names.get(actual_categories,
                                  [f'Уровень {i+1}' for i in range(actual_categories)])

        try:
            return pd.qcut(potential_series, actual_categories, labels=labels, duplicates='drop')
        except:
            bins = np.linspace(potential_series.min(), potential_series.max(), actual_categories + 1)
            return pd.cut(potential_series, bins=bins, labels=labels, include_lowest=True)

class ResultsVisualizer:
    def __init__(self, hex_gdf, features_df):
        self.merged_gdf = hex_gdf.merge(features_df, on='hex_id')

    def create_map(self, potential_scores, fitness_gdf, top_locations=None):
        """Создание интерактивной карты"""
        self.merged_gdf['potential'] = potential_scores
        centroid = self.merged_gdf.geometry.centroid.unary_union.centroid
        m = folium.Map(location=[centroid.y, centroid.x], zoom_start=11)

        merged_wgs84 = self.merged_gdf.to_crs('EPSG:4326')
        min_potential = merged_wgs84['potential'].min()
        max_potential = merged_wgs84['potential'].max()

        def get_color(value):
            normalized = (value - min_potential) / (max_potential - min_potential)
            normalized = normalized if max_potential > min_potential else 0.5

            if normalized > 0.8: return '#006400'
            elif normalized > 0.6: return '#32CD32'
            elif normalized > 0.4: return '#ADFF2F'
            elif normalized > 0.2: return '#FFFF00'
            else: return '#FFA500'

        # Добавление гексагонов
        for _, row in merged_wgs84.iterrows():
            popup_text = f"""
            <div style="font-family: Arial; font-size: 12px;">
            <h4>Информация о территории</h4>
            <b>Потенциал:</b> {row['potential']:.3f}<br>
            <b>Категория:</b> {row.get('potential_category', 'Не определена')}<br>
            <b>Кластер:</b> {row.get('cluster', 'Не определен')}<br>
            <b>ID:</b> {row['hex_id'][:12]}...
            </div>
            """

            sim_geom = row.geometry.simplify(0.001) if hasattr(row.geometry, 'simplify') else row.geometry
            folium.GeoJson(
                sim_geom,
                style_function=lambda x, row=row: {
                    'fillColor': get_color(row['potential']),
                    'color': 'black', 'weight': 0.5, 'fillOpacity': 0.7
                },
                popup=folium.Popup(popup_text, max_width=300)
            ).add_to(m)

        # Существующие фитнес-центры
        if not fitness_gdf.empty:
            fitness_wgs84 = fitness_gdf.to_crs('EPSG:4326')
            for _, row in fitness_wgs84.head(50).iterrows():
                geom_point = row.geometry.centroid if hasattr(row.geometry, 'centroid') else row.geometry
                folium.CircleMarker(
                    location=[geom_point.y, geom_point.x], radius=4,
                    popup="Существующий фитнес-центр",
                    color='red', fillColor='red', fillOpacity=0.8, weight=2
                ).add_to(m)

        if top_locations is not None and not top_locations.empty:
            top_locations_wgs84 = top_locations.to_crs('EPSG:4326')
            for _, location in top_locations_wgs84.iterrows():
                geom_point = location.geometry.centroid if hasattr(location.geometry, 'centroid') else location.geometry
                folium.Marker(
                    [geom_point.y, geom_point.x],
                    popup=folium.Popup(
                        f"<b>Рекомендуемая локация</b><br>"
                        f"<b>Потенциал:</b> {location['potential']:.3f}<br>"
                        f"<b>Категория:</b> {location.get('potential_category', 'Не определена')}",
                        max_width=250
                    ),
                    icon=folium.Icon(color='green', icon='star')
                ).add_to(m)

        self._add_legend(m, min_potential, max_potential)
        return m

    def _add_legend(self, map_obj, min_val, max_val):
        """Добавление легенды"""
        legend_html = f'''
        <div style="position: fixed; bottom: 50px; left: 50px; width: 220px;
                    background-color: white; border:2px solid grey; z-index:9999;
                    font-size:14px; padding: 10px; font-family: Arial;">
        <p style="margin: 0 0 10px 0; font-weight: bold; text-align: center;">
        Легенда потенциала</p>
        <p style="margin: 2px 0;"><i style="background: #006400; width: 20px; height: 20px;
        display: inline-block; margin-right: 5px;"></i> Очень высокий</p>
        <p style="margin: 2px 0;"><i style="background: #32CD32; width: 20px; height: 20px;
        display: inline-block; margin-right: 5px;"></i> Высокий</p>
        <p style="margin: 2px 0;"><i style="background: #ADFF2F; width: 20px; height: 20px;
        display: inline-block; margin-right: 5px;"></i> Средний</p>
        <p style="margin: 2px 0;"><i style="background: #FFFF00; width: 20px; height: 20px;
        display: inline-block; margin-right: 5px;"></i> Низкий</p>
        <p style="margin: 2px 0;"><i style="background: #FFA500; width: 20px; height: 20px;
        display: inline-block; margin-right: 5px;"></i> Очень низкий</p>
        <hr><p style="margin: 2px 0;">🔴 - Существующие фитнес-центры</p>
        <p style="margin: 2px 0;">⭐ - Рекомендуемые локации</p>
        <p style="margin: 8px 0 0 0; font-size: 12px;">
        Диапазон: {min_val:.3f} - {max_val:.3f}</p></div>
        '''
        map_obj.get_root().html.add_child(folium.Element(legend_html))

    def get_top_recommendations(self, potential_scores, n_recommendations=10, min_distance=1000):
        """Получение топ-рекомендаций"""
        self.merged_gdf['potential'] = potential_scores
        sorted_gdf = self.merged_gdf.sort_values('potential', ascending=False)

        recommendations, selected_locations = [], gpd.GeoDataFrame()
        sorted_gdf_metric = sorted_gdf.to_crs('EPSG:3857')

        for _, location in sorted_gdf_metric.iterrows():
            if len(recommendations) >= n_recommendations:
                break

            if len(selected_locations) == 0:
                recommendations.append(location)
                selected_locations = gpd.GeoDataFrame([location], crs=sorted_gdf_metric.crs)
            else:
                distances = selected_locations.geometry.distance(location.geometry)
                if distances.min() >= min_distance:
                    recommendations.append(location)
                    selected_locations = pd.concat([
                        selected_locations,
                        gpd.GeoDataFrame([location], crs=sorted_gdf_metric.crs)
                    ])

        if recommendations:
            result = gpd.GeoDataFrame(recommendations, crs=sorted_gdf_metric.crs)
            return result.to_crs(self.merged_gdf.crs)
        return gpd.GeoDataFrame(columns=self.merged_gdf.columns, crs=self.merged_gdf.crs)

class AnalysisReporter:
    """Класс для создания отчетов и визуализаций"""

    @staticmethod
    def create_comprehensive_visualization(results_df, importance_df, integral_potential,
                                         attractiveness, competition, potential_categories):
        """Создание комплексной визуализации"""
        fig = plt.figure(figsize=(20, 16))

        # Распределение потенциала
        ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
        n, bins, patches = ax1.hist(integral_potential, bins=20, alpha=0.7,
                                   color='skyblue', edgecolor='black')
        mean_potential = np.mean(integral_potential)
        ax1.axvline(mean_potential, color='red', linestyle='--', alpha=0.8,
                   label=f'Среднее: {mean_potential:.3f}')
        ax1.set_title('Распределение потенциала территорий', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Интегральный потенциал')
        ax1.set_ylabel('Количество территорий')
        ax1.grid(True, alpha=0.3)
        ax1.legend()

        # Круговая диаграмма категорий
        ax2 = plt.subplot2grid((3, 3), (0, 2))
        category_counts = pd.Series(potential_categories).value_counts()
        category_order = ['Очень низкий', 'Низкий', 'Средний', 'Высокий', 'Очень высокий']
        existing_categories = [cat for cat in category_order if cat in category_counts.index]
        category_counts = category_counts.reindex(existing_categories)

        colors = ['#FFA500', '#FFFF00', '#ADFF2F', '#32CD32', '#006400']
        ax2.pie(category_counts.values, labels=category_counts.index, autopct='%1.1f%%',
               colors=colors[:len(category_counts)], startangle=90)
        ax2.set_title('Распределение по категориям потенциала', fontsize=14, fontweight='bold')

        # Соотношение привлекательности и конкуренции
        ax3 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
        scatter = ax3.scatter(attractiveness, competition, alpha=0.6,
                             c=integral_potential, cmap='viridis', s=30)
        plt.colorbar(scatter, ax=ax3, label='Интегральный потенциал')
        ax3.set_xlabel('Привлекательность территории')
        ax3.set_ylabel('Фактор конкуренции')
        ax3.set_title('Соотношение привлекательности и конкуренции', fontsize=14, fontweight='bold')
        ax3.grid(True, alpha=0.3)

        # Анализ по кластерам
        ax4 = plt.subplot2grid((3, 3), (1, 2))
        if 'cluster' in results_df.columns:
            cluster_potential = results_df.groupby('cluster')['integral_potential'].mean()
            cluster_counts = results_df.groupby('cluster').size()

            bars = ax4.bar(range(len(cluster_potential)), cluster_potential.values,
                          color=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'][:len(cluster_potential)])
            ax4.set_xlabel('Кластер')
            ax4.set_ylabel('Средний потенциал')
            ax4.set_title('Средний потенциал по кластерам', fontsize=14, fontweight='bold')
            ax4.set_xticks(range(len(cluster_potential)))
            ax4.set_xticklabels([f'Кластер {i}' for i in cluster_potential.index])

            for i, bar in enumerate(bars):
                height = bar.get_height()
                ax4.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                        f'{height:.3f}\n({cluster_counts.iloc[i]} терр.)',
                        ha='center', va='bottom', fontsize=9)

        # Важность признаков
        ax5 = plt.subplot2grid((3, 3), (2, 0), colspan=3)
        if importance_df is not None:
            top_importance = importance_df.head(10)
            y_pos = np.arange(len(top_importance))
            bars = ax5.barh(y_pos, top_importance['importance'])

            ax5.set_yticks(y_pos)
            ax5.set_yticklabels(top_importance['feature_russian'], fontsize=10)
            ax5.set_xlabel('Важность признака')
            ax5.set_title('Топ-10 наиболее важных факторов для размещения фитнес-центров',
                         fontsize=14, fontweight='bold', pad=20)

            for i, bar in enumerate(bars):
                width = bar.get_width()
                ax5.text(width + 0.001, bar.get_y() + bar.get_height()/2,
                        f'{width:.3f}', ha='left', va='center', fontsize=10)

            ax5.grid(axis='x', alpha=0.3)

        plt.tight_layout()
        plt.show()

    @staticmethod
    def print_analysis_summary(results, top_locations, integral_potential):
        """Печать сводки анализа"""
        print("\n" + "="*60)
        print("СВОДКА АНАЛИЗА")
        print("="*60)

        print(f"Проанализировано территорий: {len(results['hex_grid'])}")
        print(f"Обнаружено фитнес-центров: {len(results['collector'].gdf_dict['fitness_centers'])}")
        print(f"Выявлено перспективных локаций: {len(top_locations)}")
        print(f"Средний потенциал: {np.mean(integral_potential):.3f}")
        print(f"Максимальный потенциал: {np.max(integral_potential):.3f}")

        if not top_locations.empty:
            print(f"\n ТОП-3 рекомендованные локации:")
            for i, (_, location) in enumerate(top_locations.head(3).iterrows(), 1):
                hex_id = str(location['hex_id'])
                hex_id_short = hex_id if len(hex_id) <= 15 else hex_id[:12] + '...'
                print(f"   {i}. Территория {hex_id_short}: потенциал = {location['potential']:.3f}")

        if 'cluster' in results['features_df'].columns:
            cluster_analysis = results['features_df'].groupby('cluster').agg({
                'integral_potential': ['mean', 'std', 'count'],
                'target': 'sum'
            }).round(3)
            print(f"\n Анализ по кластерам:")
            print(cluster_analysis)

def main():
    """Основная функция выполнения анализа"""

    print("Сбор и подготовка данных...")
    collector = GeoDataCollector("Москва")
    city_boundary = collector.get_city_boundary()
    collector.load_all_data()
    fitness_data = collector.gdf_dict["fitness_centers"]

    print("\n Создание гексагональной сетки...")
    hex_grid = HexGridBuilder(resolution=9).create_hex_grid(city_boundary)

    print("\n Инженерия пространственных признаков...")
    feature_engineer = FeatureEngineer(hex_grid, collector.gdf_dict)
    feature_engineer.calculate_spatial_features(radii=[500, 1000])
    features_df = feature_engineer.normalize_features().features_df

    print("\nОбучение моделей машинного обучения...")
    model = FitnessLocationModel()
    features_with_target = model.prepare_target_variable(features_df, fitness_data, hex_grid)
    model.train(features_with_target, features_with_target['target'])

    print("\nКластерный анализ территорий...")
    clusterer = TerritoryClusterer(max_clusters=6)
    optimal_clusters = clusterer.find_optimal_clusters(features_with_target)
    cluster_labels, _ = clusterer.perform_clustering(features_with_target, optimal_clusters)

    print("\nРасчет интегрального показателя потенциала...")
    potential_calculator = PotentialCalculator()
    integral_potential, attractiveness, competition = potential_calculator.calculate(
        hex_grid, features_with_target, model, fitness_data
    )
    potential_categories = potential_calculator.categorize(integral_potential)

    results_df = features_with_target.copy()
    results_df['cluster'] = cluster_labels
    results_df['potential_category'] = potential_categories
    results_df['integral_potential'] = integral_potential
    results_df['attractiveness_score'] = attractiveness
    results_df['competition_factor'] = competition

    print("\nВизуализация результатов...")
    visualizer = ResultsVisualizer(hex_grid, results_df)
    top_locations = visualizer.get_top_recommendations(integral_potential, n_recommendations=10)

    interactive_map = visualizer.create_map(integral_potential, fitness_data, top_locations)
    interactive_map.save('fitness_location_recommendations1.html')

    if not top_locations.empty:
        top_locations.to_file('top_recommendations.geojson', driver='GeoJSON')
    results_df.to_csv('all_location_data.csv', index=False, encoding='utf-8')

    print("Результаты сохранены:")
    print("   - fitness_location_recommendations.html (интерактивная карта)")
    print("   - top_recommendations.geojson (топ-рекомендации)")
    print("   - all_location_data.csv (все данные анализа)")

    print("\nАнализ характеристик лучших территорий...")
    importance_df = model.analyze_feature_importance()

    AnalysisReporter.create_comprehensive_visualization(
        results_df, importance_df, integral_potential,
        attractiveness, competition, potential_categories
    )

    results = {
        'collector': collector,
        'hex_grid': hex_grid,
        'features_df': results_df,
        'model': model,
        'visualizer': visualizer,
        'top_locations': top_locations,
        'integral_potential': integral_potential
    }

    AnalysisReporter.print_analysis_summary(results, top_locations, integral_potential)

    return results

if __name__ == "__main__":
    results = main()
```
